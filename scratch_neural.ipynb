```python
from matplotlib import pyplot as plt
from math import exp
import numpy as np
```

```python
inputs=[1,2,3,2.5]

weights=[[0.2,0.8,-0.5,1],
         [0.5,-0.91,0.26,-0.5],
         [-0.26,-0.27,0.17,0.87]]

biases=[2,3,0.5]
```

```python
layer_out=[]
for neuron_weight,neuron_biases in zip(weights,biases):
    neuron_output=0
    for neuron_input,weight in zip(inputs,neuron_weight):
        neuron_output+=neuron_input*weight
    neuron_output+=neuron_biases
    layer_out.append(neuron_output)
```

```python
X=[[0.2,0.8,-0.5,1],
         [0.5,-0.91,0.26,-0.5],
         [-0.26,-0.27,0.17,0.87]]
```

```python
np.random.seed(0)
class Layer_Dense:
    def __init__(self,n_inputs,n_neurons):
        self.weights=0.1*np.random.rand(n_inputs,n_neurons)
        self.biases=np.zeros((1,n_neurons))
    def forward(self,inputs):
        self.output=np.dot(inputs,self.weights)+self.biases
```

```python
layer1=Layer_Dense(4,3)
layer2=Layer_Dense(3,1)
layer3=Layer_Dense(1,1)
layer1.forward(X)
layer2.forward(layer1.output)
layer3.forward(layer2.output)
```

```python
class ActRelu:
    def forward(self,inputs):
        self.output=np.maximum(0,inputs)
```

```python
def spiral_data(points, classes):
    X = np.zeros((points*classes, 2))
    y = np.zeros(points*classes, dtype='uint8')
    for class_number in range(classes):
        ix = range(points*class_number, points*(class_number+1))
        r = np.linspace(0.0, 1, points)  # radius
        t = np.linspace(class_number*4, (class_number+1)*4, points) + np.random.randn(points)*0.2
        X[ix] = np.c_[r*np.sin(t*2.5), r*np.cos(t*2.5)]
        y[ix] = class_number
    return X, y
```

```python
x,y=spiral_data(100,3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='brg')
```

```python
layer1=Layer_Dense(2,3)
layer1.forward(x)
activation1=ActRelu()
activation1.forward(layer1.output)
print(activation1.output)
```
